# Pie chart

```python
cars=['audi','mercedes','bmw','ford','tesla','jaguar']
data=[23,17,35,29,41,12]

fig=plt.figure(figsize=(10,7))
plt.pie(data,labels=cars)

plt.show()
```

# Histogram and Binding

```python
x=[5,10,15,20,25,30,35,40,45,50]
y=[1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,'b')
plt.xlabel('x')
plt.ylabel('y')

plt.tick_params(axis='x',labelsize=18,labelrotation=-60,labelcolor='blue')
plt.tick_params(axis='y',labelsize=12,labelrotation=20,labelcolor='orange')

plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]

plt.bar(x,y)

plt.title("Bar graph")

plt.xlabel('time(hr)')
plt.ylabel('position(km)')

# plt.plot(x,y)
plt.show()
```

```python
y1=[2,3,4.5]
y2=[1,1.5,5]

plt.plot(y1)
plt.plot(y2)

plt.legend(['blue','green'],loc='lower right')

plt.show()
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()

ax=plt.axes(projection='3d')

z=np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D line Plot')

plt.show()
```
